```python
import nltk
# Define URL
url = "https://en.wikipedia.org/wiki/Les_Clark"
#Checking if URL provided contains Wikipedia
if "wikipedia" not in url:
	print("Argument not a wiki URL!")	
r = requests.get(url)
soup = BeautifulSoup(r.content,features="html5lib")
```

```python
# HTML from link
print(soup)
```

```python
article = [''.join(s.findAll(text=True)) for s in soup.findAll('p')]
article = ''.join(article)
```
